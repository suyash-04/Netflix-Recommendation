```python
import numpy as np
import pandas as pd 
```

```python
credits = pd.read_csv("../datasets/credits.csv")
credits.head()
```

```python
meta = pd.read_csv("../datasets/movies_metadata.csv")
meta.head()
```

```python
new_meta = meta.loc[meta.year  == 2017 , ['genres' , 'id' , 'title' , 'year']]
```

```python
new_meta['id']= new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits , on = 'id')
data.head()
```

```python
data['genres'][0]
```

```python
data['cast'][0]
```

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.head()
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    
    if gen == []:
        return np.nan
    else:
        return (st.join(gen))
    
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
data.head()
```

```python
data['genres_list'] 
```

```python
def get_actor(x, num):
    casts = []
    for i  in x:
        casts.append(i.get('name'))

    if casts == []:
        return np.nan
    else:
        if len(casts) < (num+1):
            return np.nan
        else:
            return (casts[num])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor(x, 0))
data['actor_2_name'] = data['cast'].map(lambda x: get_actor(x, 1))
data['actor_3_name'] = data['cast'].map(lambda x: get_actor(x, 2))
```

```python
data['actor_2_name']
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.nan
    else:
        return (st.join(dt))
```

```python

data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data.head()
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie
```

```python
movie.isna().sum()
```

```python
columns = ['director_name','actor_1_name','actor_2_name','actor_3_name']

for col in columns:
     movie[col] = movie[col].fillna("unknown")
```

```python
movie.isna().sum()
```

```python
movie = movie.dropna(how = 'any')
```

```python
movie.isna().sum()
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})

movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
movie
```
